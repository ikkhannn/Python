```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
dataset=pd.read_excel("medals.xlsx",index_col=0)
```

```python
dataset.head()
```

```python
dataset=dataset.dropna()
```

```python
fig,ax=plt.subplots(figsize=(20,5))
ax.set_xticklabels(dataset.index,rotation=90)
ax.bar(dataset.index,dataset['Gold'])
```

```python
fig,ax=plt.subplots(figsize=(20,5))
ax.bar(dataset.index,dataset['Gold'])
ax.set_xticklabels(dataset.index,rotation=90)
ax.bar(dataset.index,dataset['Silver'],bottom=dataset['Gold'])
ax.grid()
```

```python
fig,ax= plt.subplots(figsize=(20,5))
ax.set_facecolor((0.24,0.26,0.26))
ax.set_xticklabels(dataset.index,rotation=90)
ax.bar(dataset.index,dataset['Gold'],label='gold')
ax.bar(dataset.index,dataset['Silver'],bottom=dataset['Gold'],label='silver')
ax.bar(dataset.index,dataset['Bronze'],bottom=dataset['Silver'],label='bronze')
ax.grid()
ax.legend()
```
