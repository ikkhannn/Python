```python
persons=pd.DataFrame({
    'name':['Shahzad','Taha','Fatima','Khani','Zoya'],
    'age':np.random.randint(1,40,size=[5]),
    'percent':np.random.randint(40,90,size=[5])
})
```

```python
persons
```

```python
def a(age):
    if(age<18):
        return 'Non adult'
    else:
        return 'Adult'

def p(percent):
    if(percent<70):
        return 'Failed'
    else:
        return 'Passed'
```

```python
persons['Status']=persons['age'].map(a)
persons['Score']=persons['percent'].map(p)
#persons['new']=list(map(lambda p:("passed") if p>50 else "failed",persons['percent']))

persons.groupby(['Score','Status']).count()
```

```python
grouped=persons.groupby('Score')
```

```python
election_data=pd.read_csv("ElectionData2018.csv")

def resolveParty(x):
    max=0
    match=''
    for i in election_data['Party']:
        r=f.ratio(x,i)
        if(r>max):
            max=r
            match=i
    return match
```

```python
extracted_parties=election_data.Party.apply(resolveParty)
```
